```python
import requests
import re
import time  # Import the time module
from google.colab import files
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Function to extract the download link from the response text
def extract_download_link(response_text):
    match = re.search(r'window\.location="([^"]+)', response_text)
    if match:
        return match.group(1)
    return None

# Upload a txt file with URLs
uploaded = files.upload()

# Read the URLs from the uploaded file
for filename in uploaded.keys():
    with open(filename, 'r') as file:
        urls = file.readlines()

# Iterate over each URL and process it
for url in urls:
    url = url.strip()  # Clean up any whitespace or newlines
    if url:  # Check if the URL is not empty
        print(f"Processing URL: {url}")
        response = requests.get(url)

        if response.status_code == 200:
            download_link = extract_download_link(response.text)
            if download_link:
                print(f"Found download link: {download_link}")

                # Download the video using yt-dlp and save it to the specified directory
                !yt-dlp -f best -o "/content/drive/My Drive/%(title)s.%(ext)s" "{download_link}"
            else:
                print("No download link found in the response.")
        else:
            print(f"Failed to retrieve content from {url}. Status code: {response.status_code}")
```
